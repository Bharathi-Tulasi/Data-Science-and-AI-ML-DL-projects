# **Heart Disease Prediction using Statsmodels**

# Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import statsmodels.api  as sm
```

# Importing the raw data

```python
data = pd.read_csv('framingham.csv')
data
```

# Data Preprocessing

```python
data = data.dropna()
data
```

# Declare the variables

```python
y = data['TenYearCHD']
x1 = data.drop(['TenYearCHD'], axis = 1)
```

```python
x1
```

```python
x = sm.add_constant(x1)
x
```

```python
y
```

# Logistic Regression

```python
reg_log = sm.Logit(y,x).fit()
```

```python
reg_log.summary()
```

# Removing the variables which have p-value > 0.05

```python
# Removing independent variables which have p-value > 0.05
# Removing education, currentSmoker, BPMeds, prevalentStroke, prevalentHyp, diabetes, diaBP, BMI, heartRate.

copied_data = data.copy()
cleaned_data = data.drop(['education', 'currentSmoker', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'diaBP', 'BMI', 'heartRate'], axis = 1)
cleaned_data
```

# Declaring the variables

```python
Y = cleaned_data['TenYearCHD']
X1 = cleaned_data.drop(['TenYearCHD'], axis = 1)
```

```python
Y
```

```python
X = sm.add_constant(X1)
```

# Logistic Regression

```python
cleaned_reg_log = sm.Logit(Y, X).fit()
```

```python
cleaned_reg_log.summary()
```

# Accuracy

```python
pred_values = cleaned_reg_log.predict()
pred_values
```

```python
pred_values = np.round(pred_values, decimals = 0)
pred_values
```

```python
cleaned_reg_log.pred_table()
```

```python
#Confusion Matrix
cm_df = pd.DataFrame(cleaned_reg_log.pred_table())
cm_df.columns = ['Predicted 0', 'Predicted 1']
cm_df = cm_df.rename({0:'Actual 0', 1:'Actual1'})
cm_df
```

```python
# Training Accuracy
cm = np.array(cm_df)
Accuracy_train = (cm[0,0] + cm[1,1])/ (cm.sum())
Accuracy_train
```
