# Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
sns.set()
```

# Importing the Raw Data

```python
data = pd.read_csv('framingham.csv')
data
```

# Data Preprocessing

```python
data = data.dropna()
data
```

# Declaring the variables

```python
targets = data['TenYearCHD']
inputs = data.drop(['TenYearCHD'], axis = 1)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
inputs_scaled = scaler.transform(inputs)
```

# Train-Test Splitting

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 111)
```

# Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

### Predictions on the test data

```python
predictions = log_reg.predict(x_test)
```

### Confusion Matrix

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predictions)
```

```python
cm_df = pd.DataFrame(cm)
cm_df.columns = ['Predicted 0', 'Predicted 1']
cm_df = cm_df.rename({0:'Actual 0', 1:'Actual1'})
cm_df
```

### Accuracy

```python
training_accuracy = log_reg.score(x_train, y_train)
training_accuracy
```

```python
testing_accuracy = log_reg.score(x_test, y_test)
testing_accuracy
```
